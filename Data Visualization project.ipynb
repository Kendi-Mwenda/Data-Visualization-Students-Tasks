### Handling Anomalies
Outliers in the mortality rate were removed using the Interquartile Range (IQR) method to improve data reliability.

```python
Q1 = df['Mortality Rate (%)'].quantile(0.25)
Q3 = df['Mortality Rate (%)'].quantile(0.75)

IQR = Q3-Q1

outliers = df[
(df['Mortality Rate (%)'] < Q1-1.5*IQR) |
(df['Mortality Rate (%)'] > Q3+1.5*IQR)
]

print(outliers.shape)
```

### Data Transformation
Aggregation is where we combining multiple rows of data into a single summary value. These techniques were applied to compute average mortality per disease and total population affected per country. These aggregated features were merged back into the dataset to enrich the data.

```python
# Average mortality per disease
disease_avg = df.groupby('Disease Name')['Mortality Rate (%)'].mean().reset_index()
disease_avg.rename(columns={'Mortality Rate (%)': 'Avg_Disease_Mortality'}, inplace=True)
```

```python
disease_avg.head()
```

```python
# Total population affected per country
country_pop = df.groupby('Country')['Population Affected'].sum().reset_index()
country_pop.rename(columns={'Population Affected': 'Total_Pop_Affected'}, inplace=True)
```

```python
country_pop.head()
```

```python
# Merge back to main dataset
df = df.merge(disease_avg, on='Disease Name', how='left')
df = df.merge(country_pop, on='Country', how='left')
```

```python
df.head()
```

### Feature Engineering

New features such as Severity Index, Cost Burden, and Spread Intensity are created to capture deeper relationships between health, economic, and disease spread factors.

```python
df['Severity_Index'] = (df['Mortality Rate (%)'] + df['DALYs']) / df['Recovery Rate (%)'].replace(0, np.nan)
```

```python
# Cost Burden
df['Cost_Burden'] = df['Average Treatment Cost (USD)'] / df['Per Capita Income (USD)']
```

```python
# Spread Intensity
df['Spread_Intensity'] = df['Prevalence Rate (%)'] * df['Incidence Rate (%)']
```
